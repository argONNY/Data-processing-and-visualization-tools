```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify

# Загружаем данные
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/annealing/anneal.data"
cols = [f'col{i}' for i in range(1, 40)]
df = pd.read_csv(url, names=cols, header=None, na_values='?')

# Смотрим первые строки, чтобы убедиться, что col39 это классы
print(df.head(10))

# Выбираем нужные признаки
df = df[['col33','col34','col35','col39']].copy()
df.columns = ['thick','width','len','classes']

# Приводим все к строкам и убираем пробелы в классе
df['classes'] = df['classes'].astype(str).str.strip()

# Удаляем строки с пустым классом
df = df[df['classes'] != '?']
df = df[df['classes'] != 'nan']

# Преобразуем числовые колонки к float
for col in ['thick','width','len']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Заполняем пропуски средним по классу
df = df.groupby('classes').apply(lambda x: x.fillna(x.mean())).reset_index(drop=True)

# Статистика для выделенной области
thick_min, thick_median = df['thick'].min(), df['thick'].median()
width_q1, width_q3 = df['width'].quantile(0.25), df['width'].quantile(0.75)

# Area Chart
area_data = df.groupby('classes')[['thick','width','len']].mean()
plt.figure(figsize=(10,6))
area_data.plot(kind='area', alpha=0.5)
plt.title('Area Chart по классам')
plt.ylabel('Значение признака')
plt.xlabel('Классы')
plt.show()

# Treemap
sizes = df.groupby('classes')['len'].sum()
labels = sizes.index
plt.figure(figsize=(12,6))
squarify.plot(sizes=sizes, label=labels, alpha=0.7)
plt.title('Treemap по сумме len для каждого класса')
plt.axis('off')
plt.show()

# Scatter с выделенной областью
plt.figure(figsize=(8,6))
plt.scatter(df['thick'], df['width'], c='blue', alpha=0.5, label='данные')
plt.axvspan(thick_min, thick_median, color='orange', alpha=0.3, label='thick: min-медиана')
plt.axhspan(width_q1, width_q3, color='green', alpha=0.3, label='width: Q1-Q3')
plt.xlabel('thick')
plt.ylabel('width')
plt.title('Выделенная область на scatter plot')
plt.legend()
plt.show()

```
